```python
%matplotlib qt

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import os

import theano.tensor as T
import theano

from fuel.streams import DataStream

from blocks.main_loop import MainLoop
from blocks.serialization import load, load_parameters
from blocks.model import Model

from itertools import ifilter

from hft_lstm.converters import SlidingWindowPandasIterator
from hft_lstm.custom_bricks import LinearLSTM
from hft_lstm.cost import AbsolutePercentageError

import pandas as pd

import numpy as np


theano.config.floatX = 'float64'

columns_dict = {
    1: ['Close'],  # 1
    2: ['Close', 'High', 'Low'],  # 2
    3: ['Close', 'Open'],  # 3
    4: ['Close', 'High', 'Low', 'Open'],  # 4
    5: ['Close', 'Qty'],  # 5
    6: ['Close', 'Open', 'Qty'],  # 6
    7: ['Close', 'High', 'Low', 'Open', 'Qty'],  # 7
    8: ['Close', 'Vol'],  # 8
    9: ['Close', 'High', 'Low', 'Open', 'Vol'],  # 9
    10: ['Close', 'Qty', 'Vol'],  # 10
    11: ['Open', 'High', 'Low', 'Close', 'Qty', 'Vol']  # 11
} 

data_path = '.\\hft_data\\'


# Best Vale Conj.
modo = 'conj'
company = 'vale'
key = 8
hidden_dimension = 20
batch_size = 50
model_name = 'lstm_{}_{}_{}'.format(company, key, hidden_dimension)
model_location = 'C:\\Users\\tiago\\Desktop\\SAP\\saved_models\\'

# # Best Petr Todos
# modo = "todos"
# company = 'petr'
# key = 6
# hidden_dimension = 20
# batch_size = 50
# model_name = 'lstm_{}_{}_{}_{}_{}'.format(modo, company, key, hidden_dimension, batch_size)
# model_location = 'c:\\Users\\tiago\\Desktop\\AWS FODA\\hft_lstm_prediction\\saved_models\\'

# # Best Vale Todos
# modo = "todos"
# company = 'vale'
# key = 4
# hidden_dimension = 1
# batch_size = 50
# model_name = 'lstm_{}_{}_{}_{}_{}'.format(modo, company, key, hidden_dimension, batch_size)
# model_location = '.\\saved_models\\'

datafile = 'dados_{}.csv'.format(company)
# model_location = '.\\saved_models\\'
# model_location = 'c:\\Users\\tiago\\Desktop\\saved_models\\'
# model_location = 'C:\\Users\\tiago\\PycharmProjects\\hft_lstm_prediction\\saved_models\\backup_final\\'

# model_name = 'lstm_{}_{}_{}_{}_{}'.format(modo, company, key, dimension, batch_size)
# model_name = 'lstm_{}_{}_{}_{}'.format(company, key, hidden_dimension, batch_size)
# model_name = 'lstm_{}_{}_{}'.format(company, key, hidden_dimension)
```

```python
if modo == "conj":
    sw_iterator = SlidingWindowPandasIterator(data_path=data_path + datafile,
                                              normalize=True,
                                              normalize_target=False,
                                              train_size=pd.Timedelta(days=21),
                                              test_size=pd.Timedelta(days=7),
                                              step_size=pd.Timedelta(days=7))
    total = np.zeros((35,140))
else:
    sw_iterator = SlidingWindowPandasIterator(data_path=data_path + datafile,
                                              normalize=True,
                                              normalize_target=False,
                                              train_size=pd.Timedelta(days=209),
                                              test_size=pd.Timedelta(days=80),
                                              step_size=pd.Timedelta(days=7))
    if company == 'petr':
        total = np.zeros((1,1621))
    else:
        total = np.zeros((1,1624))

# train, test = next(sw_iterator)
set_window = 0

columns = columns_dict[key]

for _, test in sw_iterator:
    set_window = set_window + 1
    
    if set_window == 36:
        break
    
    print('testing window set {}\n'.format(set_window))   
    
    x_test = test[columns].values
    y_test = test[['CloseTarget']].values
    x_test1 = x_test[np.newaxis, :]
    y_test1 = y_test[:,np.newaxis]
    
    with open(model_location + model_name + '_' + str(set_window),'rb') as fparam:
        parameters = load_parameters(fparam)

    x = T.tensor3('x')
    y = T.tensor3('y')
    y = y.reshape((y.shape[1], y.shape[0], y.shape[2]))

    linear_lstm = LinearLSTM(len(columns), 1, hidden_dimension)
    y_hat = linear_lstm.apply(x)
    c = AbsolutePercentageError().apply(y, y_hat)
    # c = SquaredError().apply(y, y_hat)
    c.name = 'cost'

    model = Model(c)

    model.set_parameter_values(parameters)

    cost_matrix = list(ifilter(lambda x: x.name == 'absolutepercentageerror_cost_matrix_output_0', model.intermediary_variables))[0]
    f = theano.function([x, y], [y_hat, c, cost_matrix])
    
    y_hat_o, c_o, cm_o = f(x_test1.astype(theano.config.floatX),
                           y_test1.astype(theano.config.floatX))
    cm_o = cm_o.reshape(cm_o.shape[0])
    total[set_window-1,0:cm_o.shape[0]] = cm_o
#     for num in cm_o.reshape(cm_o.shape[0]):
#         print(str(num).replace('.',','))
```
